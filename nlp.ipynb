```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import sent_tokenize

text = "This is just random text. This is not random text"
```

```python
document = sent_tokenize(text)

type(document)
```

```python
for i in document:
    print(i) 
```

```python
from nltk.tokenize import word_tokenize

document1 = word_tokenize(text)
```

```python
for i in document1:
    print(i)
```

```python
from nltk.stem import PorterStemmer,SnowballStemmer,RegexpStemmer

pstem = PorterStemmer()

print(pstem.stem("fairly"))

snow_stem = SnowballStemmer('english') # Better than Porterstemmer.

print(snow_stem.stem('fairly'))

regex_stem = RegexpStemmer('ing$|s$|e$|able$', min=4)

print(regex_stem.stem('cutting'))

```

```python
from nltk.stem import WordNetLemmatizer

lem = WordNetLemmatizer()

print(lem.lemmatize('going',pos='n')) # pos='n' means noun
print(lem.lemmatize('going',pos='v')) # pos='v' means verb
print(lem.lemmatize('going',pos='a')) #  pos='a' means adjective
print(lem.lemmatize('going',pos='r')) # pos='r' means adverb
```
